```
small_df = df[['TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'SalePrice']].copy()
```

```
small_df.head()
```

```
small_df['TotalSquareFootage'] = small_df['TotalBsmtSF'] + small_df['1stFlrSF'] + small_df['2ndFlrSF']
small_df['PricePerSqFt'] = (small_df['SalePrice'] / small_df['TotalSquareFootage'])

small_df.head()
```

```
small_df
```

```
#run 2 regressions.
#one trying to predict whether total square footage is a good predictor of price

import plotly.express as px

fig1=px.scatter(small_df, x="TotalSquareFootage",y="SalePrice")
fig1.show()
```

```
# run a regression
import statsmodels.api as sm

x=small_df[['TotalBsmtSF','1stFlrSF','2ndFlrSF']]
y=small_df["SalePrice"]
model=sm.OLS(y,x).fit()
predictions=model.predict(x)

#summary
model.summary()
```

```
import statsmodels.api as sm

x=small_df[['TotalBsmtSF']]
y=small_df["SalePrice"]
model2=sm.OLS(y,x).fit()
predictions=model2.predict(x)

model2.summary()
```
